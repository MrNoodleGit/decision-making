```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import softmax
```

```python
values = np.array([8.7, 8.8, 8.9])
```

```python
taus = np.arange(0.01, 1, 0.001)
```

```python
softmax_list = []
for tau in taus:
    softmax_list.append(softmax(-values/tau))
    
softmax_array = np.array(softmax_list)
    
```

```python
softmax_array_transpose = softmax_array.transpose()
```

```python
softmax_array
```

```python
plt.figure(figsize=(13, 9))
for i in range(len(values)):
    plt.plot(taus, softmax_array_transpose[i])
plt.xlabel('tau')
plt.ylabel('probability')
plt.title('Probabilities as a function of tau')
plt.legend(['v1','v2','v3','v4','v5','v6','v7','v8','v9','v10'])
```
